2) Fixing Column "Installs"

```python
df['Installs'] = pd.to_numeric(
    df['Installs']
    .astype(str)
    .str.replace(r'[^0-9.-]', '', regex=True)
    .replace('', None),
    errors='coerce'
)
df
```

4) Fixing column "Price" (removed currency sympols and converted integers into strings)

```python
df['Price'] = pd.to_numeric(
    df['Price']
    .astype(str)
    .str.replace(r'[^0-9.-]', '', regex=True)
    .replace('', None),
    errors='coerce'
)
df
```

```python
df['Last Updated']=pd.to_datetime(df["Last Updated"],errors='coerce')
df
```

```python
def convert_size(size):
    if pd.isna(size):
        return None
    
    size = str(size)  # convert everything to string

    if size == "Varies with device":
        return None
    elif "M" in size:
        return float(size.replace("M", ""))
    elif "K" in size:
        return float(size.replace("K", "")) / 1024
    else:
        return None

df["Size"] = df["Size"].apply(convert_size)

print(df.head())
```

## App Dataset — Key Insights



 ###  Ratings Overview
  - **Mean:** 4.21  
  - **Median:** 4.3  
  - **Range (IQR):** 4.1 – 4.5  

  Most apps maintain **consistently high ratings**, with very few low-rated entries.  
  This indicates a **positive user bias** and generally acceptable app quality.

 

 ###  Reviews Distribution
 - **Mean:** ~444K  
 - **Median:** ~2K  
 - **Max:** ~78M  

 The distribution is **highly skewed**.  
 A small number of apps accumulate massive engagement, while the majority remain relatively unnoticed.

 > This reflects a **power-law (long-tail) behavior**.

 

 ###  Installs Pattern
 - **Mean:** ~15.4M  
 - **Median:** 100K  
 - **Max:** 1B  

 Most apps operate at a **low-to-mid install range**,  
 while a few dominate with extremely high adoption.

 > The ecosystem follows a **winner-takes-all dynamic**.

 

 ###  Pricing Structure
 - **Median Price:** 0  
 - **75% Apps:** Free  
 - **Max Price:** 400  

 The market is clearly **free-first**.  
 Paid apps form a very small fraction of the dataset.

 

 ###  Price Variability
 - **Mean Price:** 1.02  
- **Std Dev:** 15.95  

 The high standard deviation suggests **outliers in pricing**,  
 making the **median a more reliable indicator**.

 

 ###  Observed Relationships
 - Higher installs → more reviews  
 - More reviews → stable ratings  
 - High installs ≠ guaranteed high ratings  

 Popularity and quality show **partial correlation**, not direct dependence.

 

 ###  Market Summary
 - Highly competitive ecosystem  
 - Engagement concentrated among few apps  
 - Free model dominates distribution  
 - Ratings consistently skewed toward positive values
