### Filtering The Dataset

```python
df[df['Rating']>4]
```

- filtered out apps, having  rating greater than 4

```python
df[df['Category']=='FAMILY']
```

- Filterd apps that belongs to category "Family"

```python
df[df['Genres']=='Education']
```

- filterd out apps that belongs to Genre "Education"

```python
df[df['Last Updated']=='June 13, 2018']
```

- Filtered out apps that last updated on june 13 2018

```python
df[(df['Last Updated']=='June 13, 2018') & (df['Rating']>4)]
```
